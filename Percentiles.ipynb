# Charting the Data

## Percent Below Poverty

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

perc_below = df['Percent below poverty level:']
print("10% "+ str(np.percentile(perc_below,10)))
print("25% "+ str(np.percentile(perc_below,25)))
print("50% "+ str(np.percentile(perc_below,50)))
print("75% "+ str(np.percentile(perc_below,75)))
print("90% "+ str(np.percentile(perc_below,90)))
```
