# Standard Deviation
... helps look at the 'shape'... or the 'spread' of the data

## Variance && Measuring Variance
"Sigma Squared"
let peopleInLinePerHour = [1,4,5,4,8]
1. Find the mean
  - 1+4+5+4+8 = 22
  - 22 / 5 = 4.4
  - 4.4
2. Find the DIFFERENCES from the mean @ each point
  - [-3.4, -.4, .6, -.4, 3.6]
3. find the SQUARED diffs
  - [ 11.56, .16, .36, .16, 12.96]
4. find the AVG of those SQUARED Diffs
  - 11.56 + .16 + .36 + .16 + 12.96  = 25.2
  - 25.2 / 5 = 5.04
  - 5.04

...the Variance, here, is 5.04

## Standard Deviation && Measuring it
- the sq. root of variance
- sq root of 5.04 = 2.24
...the Standard Deviation, here, is 2.24
